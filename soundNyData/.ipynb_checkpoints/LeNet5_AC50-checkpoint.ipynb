```python
import torch
import torch.nn as nn
import torchvision.datasets as dataset
import torchvision.transforms as transform
from torch.utils.data import DataLoader

import numpy as np
import os

data_train = []
data_test = []

def classTrainData(): # set data_train
    path_dir = './data/train'
    file_list = os.listdir(path_dir)
    for data in file_list:
      temp = np.load('./data/train/' + data)
      tempData = torch.Tensor(temp)
      if "AD" in data:
        temp = (tempData, 0)
        data_train.append(temp)
      elif "NM" in data:
        temp = (tempData, 1)
        data_train.append(temp)
      elif "PD" in data:
        temp = (tempData, 2)
        data_train.append(temp)

def classTestData(): # set data_train
    path_dir = './data/test'
    file_list = os.listdir(path_dir)
    for data in file_list:
      temp = np.load('./data/test/' + data)
      tempData = torch.Tensor(temp)
      if "AD" in data:
        temp = (tempData, 0)
        data_test.append(temp)
      elif "NM" in data:
        temp = (tempData, 1)
        data_test.append(temp)
      elif "PD" in data:
        temp = (tempData, 2)
        data_test.append(temp)

classTrainData()
classTestData()


class LeNet5(nn.Module):
    def __init__(self):
        super(LeNet5, self).__init__()
        self.conv1 = nn.Conv2d(in_channels = 4, out_channels = 6, kernel_size = 20)
        self.conv2 = nn.Conv2d(in_channels = 6, out_channels = 8, kernel_size = 20)
        self.conv3 = nn.Conv2d(in_channels = 8, out_channels = 4, kernel_size = 20)
        self.fc1 = nn.Linear(in_features=1116, out_features=335)
        self.fc2 = nn.Linear(in_features=335,out_features=84)
        self.fc3 = nn.Linear(in_features=84,out_features=3)
        self.relu = nn.ReLU()
        self.maxPool2d = nn.MaxPool2d(kernel_size = 2, stride = 2)
        
    def forward(self, x):
        out = self.maxPool2d(x)
        out = self.maxPool2d(out)
        out = self.conv1(out)
        out = self.conv2(out)
        out = self.conv3(out)
        out = out.view(-1, 1116)
        out = self.relu(self.fc1(out))
        out = self.relu(self.fc2(out))
        out = self.fc3(out)
        return out
```

```python
batch_size = 10
learning_rate = 0.1
training_epochs = 2
loss_function = nn.CrossEntropyLoss()
network = LeNet5()
optimizer = torch.optim.Adam(network.parameters(), lr = learning_rate)
data_loader = DataLoader(dataset = data_train, batch_size=batch_size, shuffle = True, drop_last = True)

for epoch in range(training_epochs):
    avg_cost = 0
    total_batch = len(data_loader)
    
    for img, label in data_loader:
      pred = network(img)
      loss = loss_function(pred, label)
      optimizer.zero_grad()
      loss.backward()
      optimizer.step()
      avg_cost+=loss / total_batch
    print('Epoch : %d Loss = %f' %(epoch+1, avg_cost))
print('Learning finished')

# data_loader = DataLoader(dataset = data_test, batch_size= 10, shuffle = False, drop_last = True)

# count = 0
# accuracy = 0

# with torch.no_grad():
#     for img, label in data_loader: 
#       count = count + 1
#       pred = network(img)
#       correct_prediction = torch.argmax(pred, 1) == label
#       accuracy = accuracy + correct_prediction.float().mean()
#     print(accuracy / count)
```
